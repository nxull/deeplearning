```python
import numpy as np
import pandas as pd
import seaborn as sb

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_wine

from keras.models import Sequential
from keras.layers import Dense
from keras.callbacks import EarlyStopping,ModelCheckpoint
from keras.optimizers import Adamax,Adam,Adagrad
```

```python
SEED = 3
data = load_wine()
```

```python
X = data["data"]
Y = data["target"]
cols = data["feature_names"]
```

```python
df = pd.DataFrame(X,columns = cols)
df_y = pd.DataFrame(Y,columns=["target"])
```

```python
rdf = pd.concat([df,df_y],axis=1)
```

```python
sb.pairplot(rdf,vars=cols,hue="target")
```

```python
HIDDEN1 = 200
HIDDEN2 = 50
OUTPUT = 1
```

```python
X_train, X_test , Y_train, Y_test = train_test_split(X,Y,test_size=0.2,random_state=SEED)
```

```python
model = Sequential()
model.add(Dense(HIDDEN1,input_dim = X_train.shape[1],activation="relu"))
model.add(Dense(HIDDEN2,activation="relu"))
model.add(Dense(OUTPUT,activation="linear"))

model.compile(optimizer="adam",loss="mse",metrics=["accuracy"])
model.summary()
```

```python
callbacks = [EarlyStopping(monitor="val_acc",patience=20,verbose=2)]
ModelCheckpoint("checkpoints/best_model.h5",monitor="val_acc",save_best_only=True,verbose=0)
```

```python
batch_size = 64
n_epochs = 5000
model.fit(X_train,Y_train,batch_size=batch_size,epochs=n_epochs,validation_split=0.2,callbacks=callbacks)
```

```python
print("Test accuracy: %.2f" %(model.evaluate(X_test,Y_test,verbose=0)[1]))
```

## Normalize

```python
X_train_norm = StandardScaler().fit_transform(X_train)
X_test_norm = StandardScaler().fit_transform(X_test)
```

```python
model_norm = Sequential()
model_norm.add(Dense(HIDDEN1,input_dim = X_train.shape[1],activation="relu"))
model_norm.add(Dense(HIDDEN2,activation="relu"))
model_norm.add(Dense(OUTPUT,activation="linear"))

model_norm.compile(optimizer="adam",loss="mse",metrics=["accuracy"])
model_norm.summary()
```

```python
batch_size = 64
n_epochs = 5000
model_norm.fit(X_train_norm,Y_train,batch_size=batch_size,epochs=n_epochs,validation_split=0.2,callbacks=callbacks)
```

```python
print("Test accuracy: %.2f" %(model_norm.evaluate(X_test_norm,Y_test,verbose=0)[1]))
```

## Dropout

```python
from keras.layers import Dropout
```

```python
model_norm_drop = Sequential()
model_norm_drop.add(Dense(HIDDEN1,input_dim = X_train.shape[1],activation="relu"))
model_norm_drop.add(Dropout(0.2))
model_norm_drop.add(Dense(HIDDEN2,activation="relu"))
model_norm_drop.add(Dropout(0.2))
model_norm_drop.add(Dense(OUTPUT,activation="linear"))

model_norm_drop.compile(optimizer="adam",loss="mse",metrics=["accuracy"])
model_norm_drop.summary()
```

```python
batch_size = 64
n_epochs = 5000
model_norm_drop.fit(X_train_norm,Y_train,batch_size=batch_size,epochs=n_epochs,validation_split=0.2,callbacks=callbacks)
```

```python
print("Test accuracy: %.2f" %(model_norm_drop.evaluate(X_test_norm,Y_test,verbose=0)[1]))
```

```python
from keras.utils import to_categorical
from keras.datasets import mnist
from keras.utils import np_utils
import matplotlib.pyplot as plt
```

```python
(X_train, Y_train), (X_val,Y_val) = mnist.load_data()
```

```python
unique = set(Y_train)
unique
```

```python
for i,label in enumerate(unique):
    image = X_train[Y_train.tolist().index(label)]
    plt.subplot(10,10,i+1)
    plt.axis("off")
    plt.title("{0} : (({1}))".format(label,Y_train.tolist().count(label)))
    _ = plt.imshow(image,cmap="gray")
    plt.show()
```

```python
# Normalize
X_train = X_train.astype("float32")/255.
X_val = X_val.astype("float32")/255.
# One hot
Y_train = np_utils.to_categorical(Y_train,10)
Y_val = np_utils.to_categorical(Y_val,10)
# flatten data
X_train = np.reshape(X_train,(60000,784))
X_val = np.reshape(X_val,(10000,784))
```

```python
# MODEL
model_sigmoid = Sequential()
model_sigmoid.add(Dense(700,input_dim = 784,activation="sigmoid"))
model_sigmoid.add(Dense(700,activation="sigmoid"))
model_sigmoid.add(Dense(700,activation="sigmoid"))
model_sigmoid.add(Dense(700,activation="sigmoid"))
model_sigmoid.add(Dense(700,activation="sigmoid"))
model_sigmoid.add(Dense(700,activation="sigmoid"))
model_sigmoid.add(Dense(350,activation="sigmoid"))
model_sigmoid.add(Dense(100,activation="sigmoid"))
model_sigmoid.add(Dense(10,activation="sigmoid"))

model_sigmoid.compile(optimizer="sgd",loss=["categorical_crossentropy"],metrics=["accuracy"])
model_sigmoid.summary()
```

```python
# MODEL
model_relu = Sequential()
model_relu.add(Dense(700,input_dim = 784,activation="sigmoid"))
model_relu.add(Dense(700,activation="sigmoid"))
model_relu.add(Dense(700,activation="sigmoid"))
model_relu.add(Dense(700,activation="sigmoid"))
model_relu.add(Dense(700,activation="sigmoid"))
model_relu.add(Dense(700,activation="sigmoid"))
model_relu.add(Dense(350,activation="sigmoid"))
model_relu.add(Dense(100,activation="sigmoid"))
model_relu.add(Dense(10,activation="sigmoid"))

model_relu.compile(optimizer="sgd",loss=["categorical_crossentropy"],metrics=["accuracy"])

model_relu.summary()
```

```python
from keras.callbacks import Callback
class History_Loss(Callback):
    def on_train_begin(self,logs={}):
        self.losses = []
    def on_batch_end(self,batch,logs={}):
        batch_loss = logs.get("loss")
        self.losses.append(batch_loss)
```

```python
n_epoch = 10
batch_size = 256
val_split = 0.2
history_sigmoid = History_Loss()
model_sigmoid.fit(X_train,Y_train,epochs=n_epoch,batch_size=batch_size,callbacks=[history_sigmoid],validation_split=val_split)
```

```python

history_relu = History_Loss()
model_relu.fit(X_train,Y_train,epochs=n_epoch,batch_size=batch_size,callbacks=[history_relu],validation_split=val_split)
```

```python
# sb.lineplot()
#
plt.plot(np.arange(len(history_sigmoid.losses)),history_sigmoid.losses,label="sigmoid")
plt.plot(np.arange(len(history_relu.losses)),history_relu.losses,label="relu")
plt.xlabel("number of batches")
plt.ylabel("loss")
plt.title("Losses")
plt.legend(loc=1)
plt.show
```

```python
w_sigmoid = []
w_relu = []
for i in range(len(model_sigmoid.layers)):
    w_sigmoid.append(max(model_sigmoid.layers[i].get_weights()[1]))
    w_relu.append(max(model_relu.layers[i].get_weights()[1]))
```
